```python
from PIL import Image
from pylab import *
from numpy import *
import os
%matplotlib inline
```

```python
def draw_circle(w,r):
    m = arange(0,1.01,.01)*2*pi
    x = r*cos(m) + w[0]
    y = r*sin(m) + w[1]
    plot(x, y, 'b', linewidth=2)
```

```python
file = 'xiaoshizi_1.png'
image = array(Image.open(file).convert('L'))
if file[-3:] != 'pgm':
    image_1 = Image.open(file).convert('L')
    image_1.save('sif.pgm')
    image_2 ='sif.pgm'
```

```python
cmmd = str("sift "+image_2+" --output="+'eigenvalue.sift'+" "+"--edge-thresh 10 --peak-thresh 5")
os.system(cmmd)
```

```python
read_f = loadtxt('eigenvalue.sift')
descriptor = read_f[:,:4]
```

```python
figure()
gray()
imshow(image)
circle=True
for i in descriptor:
    draw_circle(i[:2], i[2])
axis('off')
show()
```
